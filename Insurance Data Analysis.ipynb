```python
# calculate the group age according to children and bmi
group_children=data.groupby("age").agg({"children":"sum","bmi":"mean"})
```

```python
# display age 
group_children
```

```python
# To find they smoke or not
data[data.smoker=="yes"]
```

```python
# Add one more column name sub region
def get_region(region):
    return region.split("h")[1]

data["subregion"]=data["region"].apply(lambda x:f"{get_region(x)}")
data.head()
```

```python
# Visualization of heatmap according to bmi and charges of dataframe
p.figure(figsize=(5,5))
sns.heatmap(group_children,annot=True)
p.title("Children According Age")
p.show()
```
